```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cleaned_V11.csv')

df.describe()
```

```python
df.info()
```

```python
# find out how many countries are represented in the dataset (in 2019)
num_countries_2019 = df[df['year'] == 2019]['country'].unique()
print(f'Number of countries represented in 2019: {len(num_countries_2019)}')
```

```python
# count nulls
null_counts = df.isnull().sum()
print("Null values in each column:\n", null_counts)

```

```python
# what countries have the most nulls
# group by country and count nulls in each group, regardless of column
nulls_by_country = df.groupby('country').apply(lambda x: x.isnull().sum().sum())
sorted_nulls_by_country = nulls_by_country.sort_values(ascending=False)
print("Null values by country:\n", sorted_nulls_by_country)
# the countries with the highest nulls seem to be the ones with the least data overall
# nulls also seem to be concentrated in "newer" countries, i.e. countries that did not exist until after colonialism / the end of the soviet union

```

```python
# create flag for countries that are in the EU
eu_countries  = ["Austria", "Belgium", "Bulgaria", "Croatia", "Cyprus", "Czechia", "Denmark", "Estonia", "Finland", "France", "Germany", "Greece", "Hungary", "Ireland", "Italy", "Latvia", "Lithuania", "Luxembourg", "Malta", "Netherlands", "Poland", "Portugal", "Romania", "Slovakia", "Slovenia", "Spain", "Sweden"]
df['is_eu'] = df['country'].isin(eu_countries)
```

```python
# make new feature with percentage of people employed
df['emp_pct'] = (df['emp'] / df['pop']) * 100
# make new feature with gdp per capita
df['rgdpna_per_capita'] = df['rgdpna'] / df['pop']
df['rgdpe_per_capita'] = df['rgdpe'] / df['pop']
df["national_total_wages"] = df['labsh'] * df['rgdpna'] 

# make feature with growth rate of gdp (per capita) over time
df['rgdpna_per_capita_pct_change'] = df.groupby('country')['rgdpna_per_capita'].pct_change() * 100
df['rgdpe_per_capita_pct_change'] = df.groupby('country')['rgdpe_per_capita'].pct_change() * 100
df['rdgpe_pct_change'] = df.groupby('country')['rgdpe'].pct_change() * 100
df['rgdpna_pct_change'] = df.groupby('country')['rgdpna'].pct_change() * 100

# It's a bit hard to determine correlation vs causation here, because of the fact that many of these features are interrelated. It's a bit of a chicken-and-egg problem, because some of these things can influence each other, especially during economic growth or recession periods.
# The influence of media in modern times, politics and social media can also play a big role in shaping behaviour, which can in turn influence economic factors like employment and wages.

# If you try to predict a recession per country, it can be simpler, but because of globalization and the limited amount of data for some countries, it can be hard to get an accurate model started. 
```

```python
# visualize gdp per capita per country 
# first, filter to only 2023 data and drop rows with null gdp or population
df_2023 = df.dropna(subset=['rgdpna', 'pop'])
df_2023 = df[df['year'] == 2023]
# filter out population < 1 million for better visualization
df_2023 = df_2023[df_2023['pop'] >= 6]

plt.figure(figsize=(12,8))
sns.scatterplot(data=df_2023, x='pop', y='rgdpe_per_capita', sizes=(1, 1000), alpha=0.7)
# log scale for better visualization
plt.yscale('log')
plt.xscale('log')
# put country names on points
for i in range(len(df_2023)):
    plt.text(x=df_2023['pop'].iloc[i], y=df_2023['rgdpe_per_capita'].iloc[i], s=df_2023['country'].iloc[i], fontsize=6)
plt.title('GDP per capita vs Population by Country in 2023')
plt.xlabel('Population')
plt.ylabel('GDP per Capita')
plt.show()
```

```python
# map all gdp's over time on one chart
plt.figure(figsize=(14,10))
#only EU countries (using is_eu flag)
sns.lineplot(data=df[df['is_eu']], x='year', y='rgdpna_per_capita', hue='country', legend="brief")
plt.title('GDP per Capita over Time for EU Countries')
plt.xlabel('Year')
plt.yscale('linear')
plt.ylabel('GDP per Capita (in current USD)')
#mark 2007 to 2009 to show financial crisis
plt.axvspan(2007, 2009, color='grey', alpha=0.3, label='Great Recession (2007-2009)')
plt.axvspan(1973, 1974, color='black', alpha=0.3, label='Oil Crisis (1973-1974)')
plt.axvspan(2019, 2022, color='lightgrey', alpha=0.3, label='Covid-19 Pandemic (2019-2022)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

```python
# map gdp per capita pct change over time for the Netherlands
plt.figure(figsize=(14,10))
sns.barplot(data=df[df['country'] == 'Netherlands'], x='year', y='rgdpna_per_capita_pct_change', color='blue')

plt.title('GDP per Capita Percentage Change over Time for the Netherlands')
#make all bars above 0 green, below 0 red
for i, bar in enumerate(plt.gca().patches):
    if bar.get_height() < 0:
        bar.set_color('red')
    else:
        bar.set_color('green')
# label every 5 years, starting from 1950
plt.xticks(ticks=np.arange(0, len(df[df['country'] == 'Netherlands']['year']), 5), labels=df[df['country'] == 'Netherlands']['year'][::5])
# add percent sign to y axis labels
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:.0f}%".format(x)))

# put grey box over 2007 to 2009 to show financial crisis
plt.axvspan(58, 60, color='grey', alpha=0.3, label='Great Recession (2007-2009)')
plt.axvspan(23, 24, color='black', alpha=0.3, label='Oil Crisis (1973-1974)')
plt.axvspan(69, 72, color='lightgrey', alpha=0.3, label='Covid-19 Pandemic (2019-2022)')


plt.xlabel('Year')
plt.ylabel('GDP per Capita Percentage Change')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
